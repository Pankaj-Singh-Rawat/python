```python
# Input: s = "abcabcbb"
# Output: 3
# Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.

s = "abcabcbb"
seen = set()
max_len = 0
for i in range(len(s)):
    if s[i] not in seen:
        seen.add(s[i])
        max_len += 1
    else:
        seen = seen.clear()

```
