```python
# Input: s = "babad"
# Output: "bab"
# Explanation: "aba" is also a valid answer.
def longestPalindrome(s: str) -> str:
    final_max = 0
    start = 0
    for i in range(len(s)):

        len1 = doThis(s, i, i)
        len2 = doThis(s, i, i + 1)

        current_max = max(len1 , len2)

        if current_max > final_max:
            final_max = current_max

            start = i - (current_max - 1)//2

    return s[start: start + final_max]

def doThis(s: str, left: int, right: int) -> int:

    while left >= 0 and right < len(s) and s[left] == s[right]:
        left -= 1
        right += 1

    return right - left - 1

s = "babad"
print(longestPalindrome(s))
```

```python
def rotate(nums: list[int], k: int) -> list:
        """
        Do not return anything, modify nums in-place instead.
        """
        k %= len(nums)
        nums.reverse()
        nums[:k] = reversed(nums[:k])
        nums[k:] = reversed(nums[k:])
        return nums

s = [1,2,3,4,5,6,7]
k = 20
print(rotate(s, k))
```

```python
# 20. Valid Parentheses
class Solution:
    def isValid(self, s: str) -> bool:
        my_dict = {')' : '(', ']' : '[', '}' : '{'}
        my_stack = []
        for char in s:
            if char in my_dict:
                last_element = my_stack.pop() if my_stack else '#'

                if my_dict[char] != last_element:
                    return False
            else:
                my_stack.append(char)
        return len(my_stack) == 0
```

```python
# 155. MinStack
class MinStack:

    def __init__(self):
        self.stack = []
        self.minStack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        if not self.minStack or self.minStack[-1] >= val:
            self.minStack.append(val)


    def pop(self) -> None:
        if self.minStack[-1] == self.stack[-1]:
            self.minStack.pop()
        self.stack.pop()


    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.minStack[-1]
```

```python
# Input: s = "3[a2[bc]]"
# Output: "accaccacc"
# "aaabcbc"
def decodeString(s: str) -> str:
    storage = ""
    num = 0
    mystack = []

    for char in s:
        if char.isdigit():
            num = num * 10 + int(char)
        elif char == '[':
            mystack.append((storage, num))
            storage = ""
            num = 0
        elif char == ']':
            last, repeat = mystack.pop()
            storage = last + (repeat * storage)

        else:
            storage += char
            
        
    return storage

s = "3[a]2[bc]"
print(decodeString(s))
```

```python
# Input: s = "(1+(4+5+2)-3)+(6+8)"
# Output: 23
# s consists of digits, '+', '-', '(', ')', and ' '.


def calculate(self, s: str) -> int:
    mystack = []
    calc = 0
    sign = 1    
    number = 0
    i = 0
    for char in s:
        if char.isdigit():
            number = number*10 + int(char)
        elif char == '+':
            calc += sign * number
            number = 0
            sign = 1
        elif char == '-':
            calc += sign * number
            number = 0
            sign = -1
        elif char == '(':  
            mystack.append((calc, sign))
            calc = 0
            sign = 1
        elif char == ')':
            calc += sign * number
            number = 0
            last_calc, last_sign = mystack.pop
            calc *= last_sign
            calc += last_calc
        
        return calc + (sign * number)
            
```

```python
# Input: nums1 = [4,1,2], nums2 = [1,3,4,2]
# Output: [-1,3,-1]

def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:
    mystack = []
    greaterElement = {}
    i = 0
    for num in nums2:

        while mystack and num > mystack[-1]:
            lesserNum = mystack.pop()
            greaterElement[lesserNum] = num

        mystack.append(num)
    return [greaterElement.get(n, -1) for n in nums1]
```

```python
def dailyTemperatures(temperatures: list[int]) -> list[int]:
# Input: temperatures = [73,74,75,71,69,72,76,73]
# Output: [1,1,4,2,1,1,0,0]

    n = len(temperatures)
    ansStack = [0] * n
    myStack = []
    for i in range(n):
        while myStack and temperatures[i] > temperatures[myStack[-1]]:
            prev_stack_index = myStack.pop()
            ansStack[prev_stack_index] = i - prev_stack_index

        myStack.append(i)

    return ansStack

temperatures = [73,74,75,71,69,72,76,73]
print(dailyTemperatures(temperatures))
```
