```python
import pandas as pd
mydataset = {
    'cars' : ['BMW', 'Meteor', 'Super'],
    'passings' : [3,1,2]
}

data1 = pd.DataFrame(mydataset)
print(data1)
```

```python
# data in series
arr = [5,4,3,2,1]

data1 = pd.Series(arr)
print(data1)
```

```python
# Labels -> index argument

print(pd.Series(arr, index = [10, 11 , 12, 13, 14]))


# key value object as series
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)
print(myvar)

# include only specific data in series
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories, index = ["day1", "day2"])
print(myvar)
```

```python
# Data Frames -> 2D
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

var1 = pd.DataFrame(data)
print(var1)

print(var1.loc[0])
print(var1.loc[[0, 2]])
```
